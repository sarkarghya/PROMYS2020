```python
n = 35
j = 3
total = 0
run = 0
for p in Primes():
    if p < 10000:
        continue
    # 1 + x^n + x^(n^2) + x^(n^3) + ... + x^(n^j)
    # to change function to x^n + x^(n^2) + x^(n^3) + ... + x^(n^j) I added #$# line code 
    used = [0 for _ in range(p)]
    for x in range(p):
        r = 0
        for ji in range(j+1):
            ji += 1 #$#
            exponent = pow(n, ji, p-1)
            r = (r+pow(x,int(exponent),p)) % p
        used[r] = 1
    goods = sum(used)/p
    total += goods
    run += 1
    print(p, '\t', numerical_approx(goods), '\t', numerical_approx(total/run))
```
